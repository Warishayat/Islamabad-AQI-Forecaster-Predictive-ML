```python
import matplotlib.pyplot as plt
import seaborn as sns

\
numeric_df = df.select_dtypes(include=['float64', 'int64'])


plt.figure(figsize=(10, 8))

sns.heatmap(
    numeric_df.corr(),
    annot=True,
    cmap='coolwarm',
    fmt=".2f",
    linewidths=0.5
)

plt.title("Air Quality & Weather Features Correlation Heatmap", fontsize=16)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
numeric_df.corr()


#There are few assumption that we can make from this table.
# 1:  Based on the correlation analysis, PM2.5 and PM10 move tightly together ($0.84$) as they represent shared particulate pollution sources, while PM2.5 exhibits a negative relationship with Ozone ($-0.33$) because clear, low-dust days facilitate higher ground-level ozone formation via solar radiation. Weather factors further regulate these levels, as higher wind speeds disperse localized air pollutants ($-0.17$) and rainfall physically washes suspended particles out of the atmosphere ($-0.04$). Finally, looking at rolling trends, recent historical averages serve as the strongest predictors of immediate future values ($0.97$), whereas higher data volatility within those short-term windows typically signals a transition into a phase of rising pollution levels ($0.44$).
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
df['Timestamp'] = pd.to_datetime(df['Timestamp'])
df['Month'] = df['Timestamp'].dt.strftime('%B')
month_order = ['January', 'February', 'March', 'April', 'May', 'June',
               'July', 'August', 'September', 'October', 'November', 'December']

plt.figure(figsize=(12, 5))
sns.barplot(x='Month', y='European_AQI', data=df, order=month_order, palette='coolwarm', errorbar=None)
plt.title("Average European AQI by Month")
plt.xticks(rotation=45)
plt.show()
```

```python
df['Hour'] = df['Timestamp'].dt.hour

plt.figure(figsize=(12, 5))
sns.lineplot(x='Hour', y='European_AQI', data=df, color='darkorange', marker='o')
plt.title("Diurnal Cycle: Average European AQI by Hour of the Day")
plt.xlabel("Hour of the Day (0-23)")
plt.ylabel("Average AQI")
plt.grid(True)
plt.xticks(range(0, 24))
plt.show()


#there is high spike from 5 to 9 becauase sun light is not much hot at that time and at the same time there is high chances ho rush on to road due to office and schools universities
```

```python
df['Day_Type'] = df['Timestamp'].dt.dayofweek.apply(lambda x: 'Weekend' if x >= 5 else 'Weekday')

plt.figure(figsize=(8, 5))
sns.boxplot(x='Day_Type', y='European_AQI', data=df, palette='Set2')
plt.title("European AQI Comparison: Weekdays vs Weekends")
plt.ylabel("AQI Score")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.scatterplot(x='Wind_Speed', y='PM2.5', data=df, alpha=0.3, color='teal')
sns.regplot(x='Wind_Speed', y='PM2.5', data=df, scatter=False, color='red')
plt.title("How Wind Speed Affects PM2.5 Concentration")
plt.xlabel("Wind Speed (m/s)")
plt.ylabel("PM2.5 Level")
plt.grid(True)
plt.show()

#if wind speed will be low pm2.5 will be high and vice versa.
```
