```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = df.sort_values("time").reset_index(drop=True)
```

```python
df.isnull().sum()
```

```python
df["pm2_5_roll_3h"] = df["pm2_5"].rolling(window=3).mean()
df["pm2_5_std_3h"] = df["pm2_5"].rolling(window=3).std()
```

```python
df["pm2_5_roll_6h"] = df["pm2_5"].rolling(window=6).mean()
df["pm2_5_std_6h"] = df["pm2_5"].rolling(window=6).std()
```

```python
df["pm2_5_roll_24h"] = df["pm2_5"].rolling(window=24).mean()
df["pm2_5_std_24h"] = df["pm2_5"].rolling(window=24).std()
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```
