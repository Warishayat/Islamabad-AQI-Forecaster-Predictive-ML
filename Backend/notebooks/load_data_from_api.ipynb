```python
import requests
import pandas as pd

def air_quality_data(lat, lon,start_date,end_date):

    url = (
        "https://air-quality-api.open-meteo.com/v1/air-quality"
        f"?latitude={lat}"
        f"&longitude={lon}"
        f"&start_date={start_date}"
        f"&end_date={end_date}"
        "&hourly=pm2_5,pm10,carbon_monoxide,nitrogen_dioxide,sulphur_dioxide,ozone,european_aqi"
    )

    response = requests.get(url)

    if response.status_code != 200:
        print("API failed:", response.status_code)
        return None

    data = response.json()

    if "hourly" not in data:
        print("No hourly key found")
        return None

    return data
```

```python
def get_weather_data(lat, lon, start_date, end_date):

    url = (
        "https://archive-api.open-meteo.com/v1/era5"
        f"?latitude={lat}"
        f"&longitude={lon}"
        f"&start_date={start_date}"
        f"&end_date={end_date}"
        "&hourly=temperature_2m,relative_humidity_2m,"
        "wind_speed_10m,pressure_msl,precipitation"
    )

    r = requests.get(url)
    if r.status_code!=200:
        print("API failed:",r.status_code)
        return None

    data = r.json()


    if "hourly" not in data:
        print("No hourly key found")
        return None
    return data
```

# time — exact timestamp of observation (hourly record)
# pm2_5 — very fine harmful particles that can directly enter the lungs
# pm10 — larger dust particles that indicate overall air pollution level
# carbon_monoxide — toxic gas mainly produced by vehicles and fuel burning
# nitrogen_dioxide — harmful gas from traffic and industrial emissions
# sulphur_dioxide — pollutant gas released from factories and burning fossil fuels
# ozone — harmful gas formed by chemical reactions between sunlight and pollutants
# european_aqi — overall air quality index representing pollution severity
# temperature_2m — air temperature measured near ground level
# relative_humidity_2m — percentage of moisture present in the air
# wind_speed_10m — speed of wind that helps disperse pollutants
# pressure_msl — atmospheric pressure indicating weather stability conditions
# precipitation — amount of rainfall that helps clean pollutants from the air
